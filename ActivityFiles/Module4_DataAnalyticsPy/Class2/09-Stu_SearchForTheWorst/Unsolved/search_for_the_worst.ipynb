```python
# Import Dependencies
import pandas as pd
from pathlib import Path
```

```python
# Create reference to CSV file
csv_file=Path("Resources/SFO_Airport_Utility_Consumption.csv")
# Import the CSV into a pandas DataFrame
pd_df = pd.read_csv(csv_file)
pd_df
```

```python
# Collect a list of all the unique values in "Utility"
utilities = pd_df["Utility"].unique()
utilities
```

```python
# Looking only at Electricity Consumption with "Tenant" owner
tenant_owner = pd_df.loc[(pd_df["Owner"] == "Tenant") & (pd_df["Utility"] == "Electricity"), :]
tenant_owner
```

```python
# Sort the DataFrame by the values in the "Usage" column to find the worst month
usage_df = tenant_owner.sort_values("Usage", ascending=False)

# Reset the index so that the index is now based on the sorting locations
new_index = usage_df.reset_index(drop=True)
new_index
```

```python
# Save all of the information collected on the worst month
worst_month = new_index.head(1)

worst_month
```
