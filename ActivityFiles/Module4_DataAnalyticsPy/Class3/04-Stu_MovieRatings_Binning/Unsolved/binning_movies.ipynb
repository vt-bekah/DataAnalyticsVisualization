```python
# Import Dependencies
import pandas as pd
from pathlib import Path
```

```python
# Create a path to the csv and read it into a Pandas DataFrame
file_path = Path("Resources\movie_scores.csv")
movies_df = pd.read_csv(file_path)
movies_df.head()
```

```python
# Figure out the minimum and maximum IMDB user vote count
min_IMDB = movies_df["IMDB_user_vote_count"].min()
print(min_IMDB)
max_IMDB = movies_df["IMDB_user_vote_count"].max()
print(max_IMDB)
```

```python
# Create bins in which to place values based upon IMDB vote count
vote_count_bins = [0, 2500, 5000, 10000, 20000, 40000, 60000, 100000, 200000, 360000]
# Create labels for these bins
vote_count_names = ["<= 2.5k", ">2.5k & <=5k",">5k & <=10k", ">10k & <=20k", ">20k & <=40k", ">40k & <=60k", ">60k & <=100k", ">100k & <=200k", ">200k & <=360k"]
```

```python
# Slice the data and place it into bins
movies_df2 = movies_df.copy()
# movies_df2["IMDB vote bin"] = pd.cut(movies_df2["IMDB_user_vote_count"], bins=vote_count_bins, labels=vote_count_names, include_lowest=True)
# movies_df2

pd.cut(movies_df2["IMDB_user_vote_count"], bins=vote_count_bins, labels=vote_count_names, include_lowest=True)
```

```python
# Place the data series into a new column inside of the DataFrame
movies_df2["IMDB vote bin"] = pd.cut(movies_df2["IMDB_user_vote_count"], bins=vote_count_bins, labels=vote_count_names, include_lowest=True)
movies_df2.head()
```

```python
# Create a GroupBy object based upon "IMDB User Votes Group"
movies_IMDBvotes = movies_df2.groupby(["IMDB vote bin"])

# Find how many rows fall into each bin

print(movies_IMDBvotes["IMDB"].count())

# Get the average of each of the first 5 rating columns within the GroupBy object
movies_IMDBvotes[["RottenTomatoes", "RottenTomatoes_User", "Metacritic", "Metacritic_User", "IMDB"]].mean()
```
