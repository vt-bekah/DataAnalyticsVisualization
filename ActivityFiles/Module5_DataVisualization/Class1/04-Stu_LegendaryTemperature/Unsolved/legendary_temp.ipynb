```python
# Dependencies
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Set x axis to numerical value for month
x_axis = np.arange(1, 13, 1)
x_axis
```

```python
# Avearge weather temp
points_F = [39, 42, 51, 62, 72, 82, 86, 84, 77, 65, 55, 44]
```

```python
# Convert to Celsius C = (F-32) * 0.56
cel = [(p-32)*0.56 for p in points_F]
cel
```

```python
# Create a handle for each plot
f_handle, = plt.plot(x_axis, points_F, marker = 'x', color = 'blue', label = "Fahrenheit")
c_handle, = plt.plot(x_axis, cel, marker = 's', color = 'red', label = "Celsius")

# Set our legend to where the chart thinks is best
plt.legend(loc='best')

# Create labels for the X and Y axis
plt.xlabel("Months")
plt.ylabel("Degrees")

# Save and display the chart
plt.savefig("../Images/TempCfg.png")
plt.show()
```
