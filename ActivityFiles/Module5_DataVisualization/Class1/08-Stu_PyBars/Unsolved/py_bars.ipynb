```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
cities = ["San Francisco", "Omaha", "New Orleans", "Cincinnati", "Pittsburgh"]
cars_in_cities = [214.7, 564.4, 416.5, 466.7, 350.6]
x_axis = np.arange(len(cars_in_cities))
```

```python
# Create a bar chart based upon the above data
plt.bar(x_axis, cars_in_cities, color='yellow', edgecolor='green', linewidth=2, alpha=.8, align="center")

# Create the ticks for our bar chart's x axis
tick_locs = [x for x in x_axis]
plt.xticks(tick_locs, cities)

# Set the limits of the x axis
plt.xlim(-0.75, len(x_axis)-0.25)

# Set the limits of the y axis
plt.ylim(0, max(cars_in_cities)*1.05)

# Give the chart a title, x label, and y label
plt.title("Density of Commuting by City")
plt.xlabel("City")
plt.ylabel("Commuting Cars Per 1k Population Age 16+")

# Save an image of the chart and print it to the screen
plt.savefig("../Images/CarDensity")
plt.show()
```
