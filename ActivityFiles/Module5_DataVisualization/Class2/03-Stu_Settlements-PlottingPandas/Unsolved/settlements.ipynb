```python
# Dependencies
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

%matplotlib inline
```

```python
# Read CSV df = pd.read_csv("../Resources/avg_rain_state.csv")
df = pd.read_csv("Resources/union_settlements_1995.csv")
df.head()
```

```python
# List columns
df.columns
```

```python
# Get total settlements by union

# what I did on my own - made an AASUMPTION what to count against
# set_by_union = df.groupby(["UNION"])["SCHEDULE NUMBER"].count().sort_values(ascending=False)

# what the solution was - this just gives the row count without counting against a particular column
set_by_union = df["UNION"].value_counts()

set_by_union

```

```python
# Configure plot, figsize, title, and axis labels
fig1 = set_by_union.plot(kind='bar', figsize=(6, 4), color='red', width=0.8, title="Major Collective Bargaining Settlements (1995)", 
                  xlabel="Union", ylabel="Settlments")

# Configure x-tick rotation plt.xticks(x, labels, rotation='vertical')
xticklabels = set_by_union.index
fig1.set_xticklabels(xticklabels, rotation=45, rotation_mode="anchor", ha="right", wrap=True)

# Show plot
plt.tight_layout()
plt.show()
```

```python
# Resize plot to display labels

# Configure plot, figsize, title, and axis labels
set_by_union.plot(kind='bar', figsize=(6, 4), color='red', width=0.8).set_xticklabels(set_by_union.index, rotation=45, rotation_mode="anchor", ha="right", wrap=True)

plt.title("Major Collective Bargaining Settlements (1995)")
plt.ylabel("Settlements")

# Configure x-tick rotation plt.xticks(x, labels, rotation='vertical')
# completed inline above with the df.plot

# Show plot
plt.tight_layout()
plt.show()
```
