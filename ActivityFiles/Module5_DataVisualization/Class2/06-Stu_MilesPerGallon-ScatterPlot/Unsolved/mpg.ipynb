```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

%matplotlib inline
```

```python
car_data = pd.read_csv('../Resources/mpg.csv')
car_data.head()
```

```python
# Remove the rows with missing values in horsepower
# car_nona = car_data.dropna()
# car_nona

# solution did this bc it wasn't na
car_data = car_data.loc[car_data["horsepower"] != "?"]
car_data
```

```python
# Set the 'car name' as our index
car_name = car_data.set_index("car name")

# Remove the 'origin' column
car_name_no = car_name.drop("origin", axis=1)

car_name_no
```

```python
# Convert the "horsepower" column to numeric so the data can be used
car_name_no["horsepower"] = pd.to_numeric(car_name_no["horsepower"])

# car_name_no["horsepower"] = car_name_no["horsepower"].astype(int64)
car_name_no["horsepower"].dtypes
```

```python
# Create a scatter plot which compares MPG to horsepower
car_name_no.plot(kind="scatter", x="horsepower", y="mpg", grid= True, figsize=(8,6), color="#1f77b4", alpha=1, title="MPG vs. Horsepower")

plt.show()
```
